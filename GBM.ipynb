```python
# Function to compute P matrix
def compute_P(grids, r, c):
    P = []

    for g in range(len(grids)):
        P.append(np.zeros((r, c), dtype=int))
        for i in range(r - 1):  # Avoid out-of-bounds for i+1
            for j in range(c - 1):  # Avoid out-of-bounds for j+1
                # Apply condition for P_{i,j}
                if grids[g][i, j] == 1 or grids[g][i, j + 1] == 1 or grids[g][i + 1, j] == 1 or grids[g][i + 1, j + 1] == 1:
                    P[g][i, j] = 1
                else:
                    P[g][i, j] = 0
    
    return P
```

```python
def solve_battleships(M, R, C, S, grids, gridsl, symGrids, r, c, s):
    
    P = compute_P(grids, r, c)

    model = gp.Model('Battleships')
    num_grids = len(grids)
    y = model.addVars(num_grids, vtype=GRB.BINARY, name="y")

    for i in range(r):
        for j in range(c):
            if T[i, j] != 100:  # Only apply constraint where T contains a clue
                model.addConstr(gp.quicksum(y[g]*symGrids[g][i, j] for g in range(num_grids))== T[i,j])

    for i in range(r):
        model.addConstr(gp.quicksum(
            y[g]*grids[g][i, j] for j in range(c)
                                for g in range(num_grids)
        ) == R[i])

    for j in range(c):
        model.addConstr(gp.quicksum(
            y[g]*grids[g][i, j] for i in range(r)
                                for g in range(num_grids)
        ) == C[j])

    for l in range(len(S)):
        model.addConstr(gp.quicksum(y[g] for g in range(s[l],s[l+1])) == S[l])

    for i in range(r):
        for j in range(c):
            model.addConstr(gp.quicksum(P[g][i, j]*y[g] for g in range(num_grids)) <= 1)

    # Solve the model
    model.setParam("Presolve", 0)
    model.optimize()

        
    # Extract solution
    if model.status == GRB.OPTIMAL:
        final_grid = np.zeros((r, c))
        selected_grids = [grids[g] for g in range(num_grids) if y[g].x > 0.5]
        for g in range(len(selected_grids)):
            final_grid += selected_grids[g]   # Sum up all selected ship placements
        return final_grid
    else:
        print("No solution found")
        return None
```

```python
rows = 10
cols = 10

# Function to determine ship symbols
def get_symbol(grid, i, j):
    if grid[i, j] == 0:
        return ""  # Water
    else:
        # Check surroundings for ship shape
        up = grid[i - 1, j] if i > 0 else 0
        down = grid[i + 1, j] if i < rows - 1 else 0
        left = grid[i, j - 1] if j > 0 else 0
        right = grid[i, j + 1] if j < cols - 1 else 0

        if left == 0 and right == 0 and up == 0 and down == 0:
            return ""  # Submarine
        elif left == 0 and right == 1:
            return ""  # Left end
        elif left == 1 and right == 0:
            return ""  # Right end
        elif up == 0 and down == 1:
            return ""  # Top end
        elif up == 1 and down == 0:
            return ""  # Bottom end
        else:
            return ""  # Middle piece



# Plot the grid
fig, ax = plt.subplots(figsize=(cols * 0.6, rows * 0.6))
ax.set_xticks(np.arange(cols + 1) - 0.5, minor=True)
ax.set_yticks(np.arange(rows + 1) - 0.5, minor=True)
ax.grid(which="minor", color="black", linewidth=2)
ax.tick_params(which="both", bottom=False, left=False, labelbottom=False, labelleft=False)

# Fill the grid with ship symbols
for i in range(rows):
    for j in range(cols):
        ax.text(j, i, get_symbol(final_grid, i, j), ha='center', va='center', fontsize=14)

# Add Row Clues
for i, clue in enumerate(R):
    ax.text(cols, i, str(clue), ha='center', va='center', fontsize=14, fontweight='bold', color='gray')

# Add Column Clues
for j, clue in enumerate(C):
    ax.text(j, rows, str(clue), ha='center', va='center', fontsize=14, fontweight='bold', color='gray')

plt.xlim(-0.5, cols + 0.5)
plt.ylim(rows + 0.5, -0.5)
plt.savefig('battleshipgrid.png')
plt.show()
```
